```python
%%writefile ambiente.py
import pandas as pd
import math
from collections import defaultdict

class AmbienteGrafo:
    def __init__(self, vertices_path, arestas_path):
        try:
            self.vertices_df = pd.read_excel(vertices_path)
            self.arestas_df = pd.read_excel(arestas_path)
        except:
            self.vertices_df = pd.read_csv(vertices_path.replace('.xlsx', '.csv'))
            self.arestas_df = pd.read_csv(arestas_path.replace('.xlsx', '.csv'))

        self.id_to_nome = dict(zip(self.vertices_df['id'], self.vertices_df['nome']))
        self.coordenadas = {}
        for _, row in self.vertices_df.iterrows():
            self.coordenadas[row['id']] = {
                'lat': row['lat'], 'lon': row['lon'], 'alt': row['altitude']
            }

        self.grafo = self._construir_grafo()
        self.n_vertices = len(self.vertices_df)

        # Estatísticas
        self.estatisticas_detalhadas = {
            'Dijkstra': {'vertices_visitados': [], 'arestas_analisadas': [], 'operacoes_fila': [], 'memoria_maxima': [], 'tempos': []},
            'A*_alpha_1': {'vertices_visitados': [], 'arestas_analisadas': [], 'operacoes_fila': [], 'memoria_maxima': [], 'tempos': []},
            'A*_alpha_0': {'vertices_visitados': [], 'arestas_analisadas': [], 'operacoes_fila': [], 'memoria_maxima': [], 'tempos': []},
            'A*_alpha_0.5': {'vertices_visitados': [], 'arestas_analisadas': [], 'operacoes_fila': [], 'memoria_maxima': [], 'tempos': []}
        }
        self.tempo_execucao = defaultdict(int) # Apenas para evitar erro de chave se chamado
        self.num_execucoes = defaultdict(int)

    def _construir_grafo(self):
        grafo = defaultdict(list)
        for _, row in self.arestas_df.iterrows():
            u, v = int(row['origem']), int(row['destino'])
            d = float(row['distancia_metros'])
            grafo[u].append((v, d))
        return grafo

    def get_vizinhos(self, v): return self.grafo.get(v, [])
    def get_nome_vertice(self, id_v): return self.id_to_nome.get(id_v, f"ID {id_v}")
    def get_coordenadas(self, id_v): return self.coordenadas.get(id_v, {'lat':0,'lon':0,'alt':0})
    def get_todos_vertices(self): return list(range(1, self.n_vertices + 1))

    def calcular_distancia_reta(self, id1, id2):
        c1, c2 = self.get_coordenadas(id1), self.get_coordenadas(id2)
        return self._haversine(c1['lat'], c1['lon'], c2['lat'], c2['lon'])

    def _haversine(self, lat1, lon1, lat2, lon2):
        lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
        dlat, dlon = lat2 - lat1, lon2 - lon1
        a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
        return 6371000 * 2 * math.atan2(math.sqrt(a), math.sqrt(1-a))

    def atualizar_estatisticas(self, algoritmo, tempo, v_vis, a_ana, ops_fila, mem_max):
        s = self.estatisticas_detalhadas[algoritmo]
        s['tempos'].append(tempo)
        s['vertices_visitados'].append(v_vis)
        s['arestas_analisadas'].append(a_ana)
        s['operacoes_fila'].append(ops_fila)
        s['memoria_maxima'].append(mem_max)
```

```python
%%writefile agentes.py
import time
import sys

class AgenteDijkstra:
    def __init__(self, ambiente):
        self.ambiente = ambiente
        self.nome = "Dijkstra"

    def encontrar_caminho(self, origem, destino):
        inicio = time.time()
        n = self.ambiente.n_vertices
        dt = {i: float('inf') for i in range(1, n + 1)}
        rot = {i: 0 for i in range(1, n + 1)}
        dt[origem] = 0

        A = set(range(1, n + 1))
        F = set()

        # Métricas
        v_vis, a_ana, ops_fila = 0, 0, 0
        mem_max = sys.getsizeof(A) + sys.getsizeof(F) + sys.getsizeof(dt)

        while F != A:
            min_dist = float('inf')
            v = None
            ops_fila += len(A)

            for vertice in A:
                if dt[vertice] < min_dist:
                    min_dist = dt[vertice]; v = vertice

            if v is None or dt[v] == float('inf'): break

            F.add(v); A.remove(v); v_vis += 1; ops_fila += 2

            for u, dist in self.ambiente.get_vizinhos(v):
                if u not in F:
                    a_ana += 1
                    if dt[v] + dist < dt[u]:
                        dt[u] = dt[v] + dist; rot[u] = v; ops_fila += 1

            mem_max = max(mem_max, sys.getsizeof(A) + sys.getsizeof(F) + sys.getsizeof(dt))

        caminho = []
        if dt[destino] < float('inf'):
            curr = destino
            while curr != 0: caminho.insert(0, curr); curr = rot[curr]

        fim = time.time()
        self.ambiente.atualizar_estatisticas(self.nome, fim-inicio, v_vis, a_ana, ops_fila, mem_max)
        return dt[destino], caminho

class AgenteAEstrela:
    def __init__(self, ambiente, alpha=0.5, fator_preguica=20.0):
        self.ambiente = ambiente
        self.alpha = alpha
        self.fator_preguica = fator_preguica
        if alpha == 1.0: self.nome = "A*_alpha_1"
        elif alpha == 0.0: self.nome = "A*_alpha_0"
        else: self.nome = f"A*_alpha_{alpha}"

    def heuristica(self, atual, destino):
        coords_a = self.ambiente.get_coordenadas(atual)
        coords_b = self.ambiente.get_coordenadas(destino)
        esforco = max(0, coords_b['alt'] - coords_a['alt'])
        preguica = self.ambiente.calcular_distancia_reta(atual, destino)
        return self.alpha * esforco + (1 - self.alpha) * preguica

    def encontrar_caminho(self, origem, destino):
        inicio = time.time()
        F = set()
        A = {origem}
        g = {i: float('inf') for i in range(1, self.ambiente.n_vertices + 1)}
        f = {i: float('inf') for i in range(1, self.ambiente.n_vertices + 1)}
        rot = {i: 0 for i in range(1, self.ambiente.n_vertices + 1)}

        g[origem] = 0
        f[origem] = self.heuristica(origem, destino)

        v_vis, a_ana, ops_fila = 0, 0, 0
        mem_max = sys.getsizeof(A) + sys.getsizeof(g) + sys.getsizeof(f)

        while A:
            atual = min(A, key=lambda x: f[x])
            ops_fila += len(A)

            if atual == destino: break

            A.remove(atual); F.add(atual); v_vis += 1; ops_fila += 2

            for v, dist in self.ambiente.get_vizinhos(atual):
                if v in F: continue
                a_ana += 1

                #LÓGICA DO FATOR PREGUIÇA
                c_atual = self.ambiente.get_coordenadas(atual)
                c_viz = self.ambiente.get_coordenadas(v)
                subida = max(0, c_viz['alt'] - c_atual['alt'])
                custo_real = g[atual] + dist + (subida * self.fator_preguica)

                if v not in A or custo_real < g[v]:
                    rot[v] = atual
                    g[v] = custo_real
                    f[v] = g[v] + self.heuristica(v, destino)
                    if v not in A: A.add(v); ops_fila += 1
                    else: ops_fila += 1

            mem_max = max(mem_max, sys.getsizeof(A) + sys.getsizeof(g) + sys.getsizeof(f))

        # Reconstrói caminho e calcula distância física real (sem a penalidade)
        caminho = []
        dist_fisica = 0
        if g[destino] < float('inf'):
            curr = destino
            while curr != 0:
                caminho.insert(0, curr)
                pai = rot[curr]
                if pai != 0:
                    for viz, d in self.ambiente.get_vizinhos(pai):
                        if viz == curr: dist_fisica += d; break
                curr = pai

        fim = time.time()
        self.ambiente.atualizar_estatisticas(self.nome, fim-inicio, v_vis, a_ana, ops_fila, mem_max)
        return dist_fisica, caminho
```

```python
%%writefile relatorios.py
import matplotlib.pyplot as plt
import numpy as np

class ImprimeResultados:
    @staticmethod
    def gerar_relatorio_completo(resultados, ambiente):
        filename = 'comparacao_caminhos_completo.txt'
        with open(filename, 'w', encoding='utf-8') as f:
            f.write("=" * 100 + "\n")
            f.write("COMPARAÇÃO DE CAMINHOS ENCONTRADOS POR CADA ALGORITMO\n")
            f.write("=" * 100 + "\n\n")

            f.write(f"Total de pares de vértices testados: {len(resultados)}\n")
            f.write(f"Número total de vértices no grafo: {ambiente.n_vertices}\n\n")

            for item in resultados:
                f.write("=" * 80 + "\n")
                f.write(f"PAR {item['id_par']}: ID: {item['origem']} -> ID: {item['destino']}\n")
                f.write(f"Rota: {ambiente.get_nome_vertice(item['origem'])} -> {ambiente.get_nome_vertice(item['destino'])}\n")
                f.write("-" * 80 + "\n")

                def formatar_caminho(lista_ids):
                    nomes = [ambiente.get_nome_vertice(uid) for uid in lista_ids]
                    return f"{lista_ids}  ({ ' -> '.join(nomes) })"

                def escrever_bloco(titulo, dados):
                    f.write(f"{titulo}:\n")
                    f.write(f"   Distância: {dados['dist']:.2f} m\n")
                    f.write(f"   Caminho: {formatar_caminho(dados['path'])}\n\n")

                escrever_bloco("1. DIJKSTRA", item['dijkstra'])
                escrever_bloco("2. A* (alpha=1.0) [Foca no Relevo]", item['a1'])
                escrever_bloco("3. A* (alpha=0.0) [Foca na Distância]", item['a0'])
                escrever_bloco("4. A* (alpha=0.5) [Híbrido]", item['a05'])

                f.write("=" * 80 + "\n\n")

        print(f"✓ Arquivo '{filename}' gerado com sucesso!")

    @staticmethod
    def gerar_txt_divergencias(divergencias, ambiente):
        with open('relatorio_divergencias.txt', 'w', encoding='utf-8') as f:
            f.write("=" * 100 + "\n")
            f.write(f"{'RELATÓRIO DE DIVERGÊNCIAS (RESUMO)':^100}\n")
            f.write("=" * 100 + "\n\n")
            for d in divergencias:
                f.write(f"ROTA: {d['rota']}\n")
                f.write(f"Dijkstra: {d['d_dist']:.0f}m  |  A* (1.0): {d['a1_dist']:.0f}m\n")
                f.write("-" * 50 + "\n")

    @staticmethod
    def gerar_tabela_performance(estatisticas):
        """Gera e imprime a tabela de médias de desempenho no terminal e salva em TXT"""

        # Filtra algoritmos que rodaram
        algoritmos = [alg for alg in estatisticas.keys() if estatisticas[alg]['tempos']]
        if not algoritmos: return

        linhas = []
        linhas.append("=" * 115)
        linhas.append(f"{'TABELA DE DESEMPENHO MÉDIO (ESTATÍSTICAS)':^115}")
        linhas.append("=" * 115)

        # Cabeçalho da Tabela
        header = f"{'Algoritmo':<20} | {'Tempo (ms)':<12} | {'Vértices':<10} | {'Arestas':<10} | {'Ops Fila':<10} | {'Memória (B)':<12}"
        linhas.append(header)
        linhas.append("-" * 115)

        # Preenche os dados
        for alg in algoritmos:
            stats = estatisticas[alg]

            # Calcula médias
            tempo_ms = np.mean([t * 1000 for t in stats['tempos']])
            vertices = np.mean(stats['vertices_visitados'])
            arestas = np.mean(stats['arestas_analisadas'])
            fila = np.mean(stats['operacoes_fila'])
            memoria = np.mean(stats['memoria_maxima'])

            # Formata nome para caber na tabela
            nome_clean = alg.replace("A*_alpha_", "A* α=")

            row = f"{nome_clean:<20} | {tempo_ms:>10.2f}   | {vertices:>10.1f} | {arestas:>10.1f} | {fila:>10.1f} | {memoria:>12.0f}"
            linhas.append(row)

        linhas.append("-" * 115)

        # Imprime no terminal
        print("\n".join(linhas))

        # Salva em arquivo para não perder
        with open('tabela_desempenho.txt', 'w', encoding='utf-8') as f:
            f.write("\n".join(linhas))
            print("\n✓ Tabela salva em 'tabela_desempenho.txt'")

    @staticmethod
    def imprimir_graficos(stats):
        print("Gerando gráficos comparativos...")
        algoritmos = ['Dijkstra', 'A*_alpha_1', 'A*_alpha_0', 'A*_alpha_0.5']
        labels = ['Dijkstra', 'A* (α=1)\nRelevo', 'A* (α=0)\nDistância', 'A* (α=0.5)\nHíbrido']
        cores = ['blue', 'red', 'green', 'purple']

        fig, axes = plt.subplots(2, 2, figsize=(16, 12))
        fig.suptitle("Comparativo de Desempenho (Médias por Rota)", fontsize=16)

        metricas = [
            ('tempos', 'Tempo Médio (s)', 1000),
            ('vertices_visitados', 'Vértices Visitados (méd)', 1),
            ('arestas_analisadas', 'Arestas Analisadas (méd)', 1),
            ('operacoes_fila', 'Ops na Fila (méd)', 1)
        ]

        for i, ax in enumerate(axes.flatten()):
            chave_dict, titulo, fator = metricas[i]
            vals = []
            for alg in algoritmos:
                if alg in stats and len(stats[alg][chave_dict]) > 0:
                    media = np.mean(stats[alg][chave_dict]) * fator
                    vals.append(media)
                else:
                    vals.append(0)

            bars = ax.bar(labels, vals, color=cores, alpha=0.8, edgecolor='black')
            ax.set_title(titulo, fontsize=12, fontweight='bold')
            ax.grid(axis='y', linestyle='--', alpha=0.3)

            for bar in bars:
                height = bar.get_height()
                texto = f'{height:.4f}' if i == 0 else f'{height:.1f}'
                ax.text(bar.get_x() + bar.get_width()/2., height, texto, ha='center', va='bottom', fontsize=10)

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()
```

```python
%%writefile main.py
from ambiente import AmbienteGrafo
from agentes import AgenteDijkstra, AgenteAEstrela
from relatorios import ImprimeResultados
import warnings

warnings.filterwarnings('ignore')

def executar():
    print(">>> INICIANDO SIMULAÇÃO COMPLETA <<<")

    try:
        amb = AmbienteGrafo('vertices.xlsx', 'arestas.xlsx')
        print(f"✓ Mapa carregado: {amb.n_vertices} vértices.")
    except Exception as e:
        print(f"ERRO: Arquivos Excel não encontrados ou inválidos. {e}")
        return

    # FATOR 20.0: pra forçar o desvio do morro
    FATOR = 20.0

    print(">>> Instanciando Agentes (Fator Preguiça 20.0)...")
    dijkstra = AgenteDijkstra(amb)
    astar_1   = AgenteAEstrela(amb, alpha=1.0, fator_preguica=FATOR)
    astar_0   = AgenteAEstrela(amb, alpha=0.0, fator_preguica=FATOR)
    astar_05  = AgenteAEstrela(amb, alpha=0.5, fator_preguica=FATOR)

    todos_vertices = amb.get_todos_vertices()
    pares = [(u, v) for u in todos_vertices for v in todos_vertices if u != v]

    print(f"✓ Testando {len(pares)} rotas...")

    # INICIALIZAÇÃO DAS LISTAS
    caminhos_completos = []
    divergencias = []

    for i, (origem, destino) in enumerate(pares):
        if i % 50 == 0: print(f"   Processando... {i}/{len(pares)}")

        d_dist, path_d     = dijkstra.encontrar_caminho(origem, destino)
        a1_dist, path_a1   = astar_1.encontrar_caminho(origem, destino)
        a0_dist, path_a0   = astar_0.encontrar_caminho(origem, destino)
        a05_dist, path_a05 = astar_05.encontrar_caminho(origem, destino)

        caminhos_completos.append({
            'id_par': i + 1,
            'origem': origem,
            'destino': destino,
            'dijkstra': {'dist': d_dist, 'path': path_d},
            'a0':       {'dist': a0_dist, 'path': path_a0},
            'a05':      {'dist': a05_dist, 'path': path_a05},
            'a1':       {'dist': a1_dist, 'path': path_a1}
        })

        if (d_dist != a05_dist) or (d_dist != a0_dist) or (d_dist != a1_dist):
            divergencias.append({
                'rota': f"{amb.get_nome_vertice(origem)} -> {amb.get_nome_vertice(destino)}",
                'd_dist': d_dist,
                'a0_dist': a0_dist,
                'a05_dist': a05_dist,
                'a1_dist': a1_dist
            })

    print(f"\n✓ Simulação concluída.")
    print(f"✓ Rotas Divergentes: {len(divergencias)}")

    ImprimeResultados.gerar_relatorio_completo(caminhos_completos, amb)
    ImprimeResultados.gerar_txt_divergencias(divergencias, amb)
    ImprimeResultados.gerar_tabela_performance(amb.estatisticas_detalhadas)
    ImprimeResultados.imprimir_graficos(amb.estatisticas_detalhadas)

if __name__ == "__main__":
    executar()
```
